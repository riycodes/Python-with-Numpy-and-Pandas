their sum
their average (mean)
the biggest one

```python
import numpy as np
scores = np.array([72, 45, 88, 91, 38, 67, 55, 80])
print(np.sum(scores))
print(np.mean(scores))
print(np.max(scores))
print(scores[4:8])
print(scores[:4])
print(scores[-4:])
print(scores[0:-4])
```

SHAPES: When you get a real dataset, the first thing you ask is "how big is it?" — how many rows, how many columns? NumPy answers with .shape.
Notice something new: there's no np. and no brackets here. shape isn't a tool you call — it's a piece of information attached to the array itself. You ask the array about itself

```python
print(np.shape(scores)) #1d array, counts item in the array, then leaves column side empty. [row, column]
```

Meeting the Series container of Pandas

A Series is like a NumPy array, but with one upgrade: each value has a label. A plain array just has positions (0, 1, 2...). A Series has positions and names you choose.

```python
marks = pd.Series([85,90,70])
print(marks)
```

```python
import numpy as np
marks = np.array([85,90,70])
print(marks)
```

```python
prices = pd.Series([100,250,80], index=["Pen","Bag", "Book"])
print(prices)
```
